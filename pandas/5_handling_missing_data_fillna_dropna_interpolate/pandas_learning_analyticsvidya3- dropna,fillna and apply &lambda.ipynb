# dropna,fillna and apply &lambda

```python
import numpy as np
import pandas as pd
from pandas import read_csv
```

```python
filename='residence.csv'
header_names=["hosue no","apartment no","city","Family available","no of years","actual years","rent"]
data=read_csv(filename,names=header_names)
```

```python
data
```

# dropna() method manage and remove Null/NAN/NA values from a data frame, 


# fillna() manages and let the user replace NaN values with some value of their own.

```python
data["rent"].fillna("500",inplace=True)
```

```python
data
```

# bfill, backfill and pad methods also available

```python
#method is set as ffill and hence the value in the same column replaces the null value.
#a limit of 1 is set in the fillna() method to check if the function stops replacing after 
#one successful replacement of NaN value or not.

data["city"].fillna(method="ffill",limit=1)
data
```

```python
new_data=data.dropna(axis=0,how="any")
new_data
```

```python
#how: how takes string value of two kinds only (‘any’ or ‘all’). 
#‘any’ drops the row/column if ANY value is Null and ‘all’ drops only if ALL values are null.


new_data=data.dropna(axis=1,how="any")
new_data
```

```python
#Use apply funtions as applying conditions on individual row or a column in a dataframe
def clip_rent(rent_price):
    if rent_price > 207:
        rent_price = 205
    return rent_price
#after clipping
#data["apartment no"].apply(lambda x: clip_rent(x))     -->with lambda
data["apartment no"].apply(clip_rent)    #without lambda
```

```python
#If i should need data in 1 and 0 format
def label_encode(city_name) :
    if city_name =='Bangalore' :
        label="Bangalore"
    elif city_name =="NewDelhi" :
        label="NewDelhi"
    else :
        label="NA"
    return label

#operate label_encode in every row of city

data["city"]=data["city"].apply(label_encode)
data["city"]
```

```python
def label_encode(city_name) :
    if city_name =='Bangalore' :
        label=0
    elif city_name =="NewDelhi" :
        label=1
    else :
        label=2
    return label

#operate label_encode in every row of city

data["city"]=data["city"].apply(label_encode)
data["city"]
```
